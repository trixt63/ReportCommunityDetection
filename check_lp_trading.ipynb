```python
import pandas as pd
import seaborn as sns
```

```python
file = open('./artifacts/0x38_dextools_oldest_transactions.txt')
_month_names = ['Jun', 'Jul', "Aug", "Sep", "Oct", "Nov", "Dec", "Jan", "Feb", "Mar", "Apr", "May"]
n_lp_months = {month_name: 0 for month_name in _month_names}
n_lp_may = [0]*31
while True:
    contract_oldest_tx = file.readline()
    if not contract_oldest_tx:
        break
    contract_addr = contract_oldest_tx.split(' ')[0][:-1]
    tx_month = contract_oldest_tx.split(' ')[2]
    tx_day = contract_oldest_tx.split(' ')[3]
    tx_time = contract_oldest_tx.split(' ')[4]

    n_lp_months[tx_month] += 1

    if tx_month == 'May':
        n_lp_may[int(tx_day) - 1] += 1
    a = 0

print(f"Number of last tx each mont: {n_lp_months}")
print(f"Number of last transaction each day in May: {n_lp_may}")
```

```python
months_df = pd.DataFrame({'month': n_lp_months.keys(),  'number of pairs': n_lp_months.values()})
print(months_df)
sns.set(rc={'figure.figsize': (12, 6)})
sns.set_palette('muted')
ax = sns.barplot(data=months_df, x="month", y="number of pairs")
for i in ax.containers:
    ax.bar_label(i)
```

```python
days_df = pd.DataFrame({'day': range(1, 32), 'number of pairs': n_lp_may})
# print(days_df)
sns.set(rc={'figure.figsize': (12, 6)})
ax = sns.barplot(data=days_df, x="day", y="number of pairs")
for i in ax.containers:
    ax.bar_label(i)
```
